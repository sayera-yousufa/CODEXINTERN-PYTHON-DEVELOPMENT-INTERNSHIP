```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set visual style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Load Titanic dataset
df = pd.read_csv("C:/Users/hp/Downloads/train.csv")

# Basic info
print("First 5 rows:\n", df.head())
print("\nData Summary:\n", df.describe(include='all'))
print("\nMissing Values:\n", df.isnull().sum())

# Drop rows with missing 'Age' for clean analysis (optional)
df_clean = df.dropna(subset=['Age'])

# Average Age
avg_age = df_clean['Age'].mean()
print(f"\nAverage Age of Passengers: {avg_age:.2f} years")

# Average Fare
avg_fare = df['Fare'].mean()
print(f"Average Fare: ${avg_fare:.2f}")
```

```python
# Bar Chart: Survival Count
sns.countplot(x='Survived', data=df)
plt.title('Survival Count (0 = Not Survived, 1 = Survived)')
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')
plt.show()

# Bar Chart: Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival Count by Gender')
plt.show()
```

```python
# Scatter Plot: Age vs Fare
plt.scatter(df_clean['Age'], df_clean['Fare'], alpha=0.5)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
# Heatmap: Correlation
corr_matrix = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Some simple insights
print("\nInsights:")
print("- Younger passengers paid more on average.")
print("- Women had a higher survival rate than men.")
print("- Higher Pclass (1st class) passengers were more likely to survive.")
```
